```python
import numpy as np
import matplotlib.pyplot as plt

# Dados tabulados
categories = ["Fan", "LPC", "HPC", "HPT", "LPT", "VSV", "VBV", "Nf", "Nc", "P24", "Ps30",
              "T24", "T30", "T48", "WF36", "P2", "T2", "Pamb", "No Fault"]

ann2_values = np.array([82, 27, 86, 96, 88, 83, 55, 72, 27, 69, 58, 43, 53, 69, 81, 68, 67, 71, 99.18])
cnn2_values = np.array([71, 21, 74, 94, 75, 81, 45, 58, 22, 71, 46, 39, 51, 71, 79, 64, 61, 2, 97.23])

x = np.arange(len(categories))  # Posições das categorias

# Criando o gráfico de barras sobrepostas
fig, ax = plt.subplots(figsize=(12, 5))
bar_width = 0.6

# Barras para CNN2 (maiores)
ax.bar(x, cnn2_values, bar_width, label="CNN 2", color="blue", edgecolor="black")

# Barras para ANN2 (dentro da outra, menores)
ax.bar(x, ann2_values, bar_width * 0.6, label="ANN 2", color="black", edgecolor="black")

# Configuração dos eixos
ax.set_xlabel("Componentes e Sensores", fontsize=12)
ax.set_ylabel("CCR (%)", fontsize=12)
ax.set_title("Comparação de CCR (%) entre ANN 2 e CNN 2", fontsize=14)
ax.set_xticks(x)
ax.set_xticklabels(categories, rotation=45, ha="right")
ax.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Novos dados tabulados
categories = ["Fan", "LPC", "HPC", "HPT", "LPT", "VSV", "VBV", "Nf", "Nc", "P24", "Ps30",
              "T24", "T30", "T48", "WF36", "P2", "T2", "Pamb", "No Fault"]

ann2_values = np.array([67, 35, 69, 83, 71, 54, 29, 44, 27, 47, 53, 42, 38, 44, 56, 42, 47, 62, 99.18])
cnn2_values = np.array([55, 34, 63, 81, 66, 52, 24, 38, 22, 46, 47, 41, 39, 41, 57, 36, 33, 1, 97.23])

x = np.arange(len(categories))  # Posições das categorias

# Criando o gráfico de barras sobrepostas
fig, ax = plt.subplots(figsize=(12, 5))
bar_width = 0.6

# Barras para CNN2 (maiores)
ax.bar(x, cnn2_values, bar_width, label="CNN 2", color="blue", edgecolor="black")

# Barras para ANN2 (dentro da outra, menores)
ax.bar(x, ann2_values, bar_width * 0.6, label="ANN 2", color="black", edgecolor="black")

# Configuração dos eixos
ax.set_xlabel("Componentes e Sensores", fontsize=12)
ax.set_ylabel("CCR (%)", fontsize=12)
ax.set_title("Comparação de CCR (%) entre ANN 2 e CNN 2 para falhas rápidas", fontsize=14)
ax.set_xticks(x)
ax.set_xticklabels(categories, rotation=45, ha="right")
ax.legend()

plt.tight_layout()
plt.show()
```
